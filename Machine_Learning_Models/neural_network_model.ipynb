```python
# Step 1: Import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import tensorflow as tf

# Step 2: Load the cleaned merged dataset
df = pd.read_csv("D:/diabetes-risk-prediction-model/Resources/full_cleaned_merged.csv")

# Step 3: Drop missing target values
df = df.dropna(subset=["Diabetes_binary"])
df = df.reset_index(drop=True)

# Step 4: Split features and target
X = df.drop(columns=["Diabetes_binary"])
y = df["Diabetes_binary"]

# Step 5: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 6: Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Step 7: Build the Neural Network model
model = tf.keras.models.Sequential()

# Input layer + First hidden layer
model.add(tf.keras.layers.Dense(units=64, activation='relu', input_shape=(X_train_scaled.shape[1],)))

# Second hidden layer
model.add(tf.keras.layers.Dense(units=32, activation='relu'))

# Output layer for binary classification (diabetic or not)
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Step 8: Train the model
history = model.fit(
    X_train_scaled,
    y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

```python
# Step 9: Evaluate the model
loss, accuracy = model.evaluate(X_test_scaled, y_test, verbose=2)

print("Neural Network Accuracy on Test Data:", accuracy)
```
