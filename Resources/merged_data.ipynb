```python
import pandas as pd

# Load all three cleaned files
df1 = pd.read_csv("diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
df2 = pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")
df3 = pd.read_csv("diabetes_binary_health_indicators_BRFSS2015.csv")

# Check their shapes (rows, columns)
print("File 1:", df1.shape)
print("File 2:", df2.shape)
print("File 3:", df3.shape)
```

```python
# Merge all 3 cleaned datasets into one big dataset
combined_df = pd.concat([df1, df2, df3], ignore_index=True)

# Check the shape of the combined data
print("Combined data shape:", combined_df.shape)

# Preview first few rows
combined_df.head()
```

```python
# Drop the unwanted column from File 2
combined_df = combined_df.drop(columns=["Diabetes_012"])

# Check the new shape
print("Cleaned merged data shape:", combined_df.shape)
```

```python
# Save the combined cleaned data to a CSV file
combined_df.to_csv("full_cleaned_merged.csv", index=False)

print("✅ File saved successfully as full_cleaned_merged.csv")
```
