```python
import pandas as pd
```

```python
df_012 = pd.read_csv('Recourcediabetes_012_health_indicators_BRFSS2015.csv')
df_012.head()
```

```python
df_012.info()
```

```python
missing_values = df_012.isnull().sum()
missing_values
```

```python
duplicates = df_012.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Remove duplicate rows
df_012 = df_012.drop_duplicates()
```

```python
# Identify likely categorical columns
categorical_cols = [
    'Diabetes_012', 'HighBP', 'HighChol', 'CholCheck', 'Smoker', 'Stroke',
    'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies',
    'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'DiffWalk',
    'Sex', 'GenHlth', 'Education', 'Income'
]

# Convert them to 'category' type
for col in categorical_cols:
    df_012[col] = df_012[col].astype('category')

# Verify data types again
print("Updated data types:")
print(df_012.dtypes)
```

```python
df_012.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot class distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Diabetes_012', data=df_012)
plt.title('Distribution of Diabetes Classes')
plt.xlabel('Diabetes Status')
plt.ylabel('Count')
plt.show()

# Print counts
print(df_012['Diabetes_012'].value_counts())
```

```python
# Example: Age by diabetes status
plt.figure(figsize=(8,6))
sns.boxplot(x='Diabetes_012', y='Age', data=df_012)
plt.title('Age Distribution by Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Diabetes_012', y='BMI', data=df_012)
plt.title('BMI Distribution by Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('BMI')
plt.show()
```

```python
phys_activity_rates = df_012.groupby('PhysActivity')['Diabetes_012'].value_counts(normalize=True).unstack()
print(phys_activity_rates)
phys_activity_rates.plot(kind='bar', stacked=True)
plt.title('Diabetes Status by Physical Activity')
plt.xlabel('Physical Activity (0=No, 1=Yes)')
plt.ylabel('Proportion')
plt.show()
```

```python
bp_diabetes = pd.crosstab(df_012['HighBP'], df_012['Diabetes_012'], normalize='index')
print(bp_diabetes)
bp_diabetes.plot(kind='bar', stacked=True)
plt.title('Diabetes Status by High Blood Pressure')
plt.xlabel('High Blood Pressure (0=No, 1=Yes)')
plt.ylabel('Proportion')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Diabetes_012', y='Age', data=df_012)
plt.title('Age Distribution by Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='GenHlth', hue='Diabetes_012', data=df_012)
plt.title('General Health vs. Diabetes Status')
plt.xlabel('General Health (1=Excellent, 5=Poor)')
plt.ylabel('Count')
plt.show()
```
